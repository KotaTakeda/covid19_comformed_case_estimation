# データを視覚化

```python
import pandas as pd
```

```python
grand_total_csv_path = 'who_situation_reports/csv/masters/grand_total.csv'
grand_total_df = pd.read_csv(grand_total_csv_path, index_col=0)
grand_total_df
```

```python
ten_df = grand_total_df.loc[~grand_total_df.isnull().any(axis=1)].iloc[7:,:]
ten_df
x  = ten_df['Total conformed new cases']
y = ten_df.index.values

print(x)
print(y)
%matplotlib inline
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(50,20))
plt.plot(x,y)
```
