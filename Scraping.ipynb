## サイトにアクセスしてscraping

```python
who_sitrep_url = 'https://www.who.int/emergencies/diseases/novel-coronavirus-2019/situation-reports'

scraping_result = scraping_href(who_sitrep_url)
```

## pdfのリンクを抽出

```python
urls_for_pdf = extract_pdf_urls(scraping_result)
```

## データをまとめるマスターのtable作成


```python
# とりあえず全世界のdataframeを作る
master = pd.DataFrame(columns = REPOERT_COLUMNS[1:5])
grand_total_csv_path = 'who_situation_reports/csv/masters/grand_total.csv'
master.to_csv(grand_total_csv_path)

# todo:
# 'Total conformed cases'
# total_comf = pd.Data

# 'Total conformed new cases'
# 'Total deaths'
# 'Total new deaths'
# 'Days since last reported case'
```

## pdfダウンロード


```python
# 1.pdfをダウンロード

# 実行
N = len(urls_for_pdf)
for i in range(N):
    print('================================')
    print(i)
    print('================================')

    # 古い順にダウンロード
#     url = urls_for_pdf[i]
    url = urls_for_pdf[-(i+1)]
    
    # pdfダウンロード
    save_pdf_path = download_pdf(url)
    
    # pdf -> csv
    save_csv_path = pdf_into_csv(save_pdf_path)
    
    # grand_totalに書き込み
    written = write_to_master_csv(save_csv_path, grand_total_csv_path)
    print(written)
    

#  csv整形 todo
#     各日によってdfのindex, columnsの数が変化するのでcleaned_dfをするのは延期
#     cleaned_df = cleaning_csv(save_csv_path)
#     write_to_master_csv(save_csv_path, master_path)


```
